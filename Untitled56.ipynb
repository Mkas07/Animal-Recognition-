```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the trained model
model = tf.keras.models.load_model('/content/cats_dogs_pandas_modelfinal.h5')
```

```python
from ipywidgets import FileUpload
from IPython.display import display
import matplotlib.pyplot as plt
# Preprocess the input image
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(150, 150))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  # Rescale the image to [0, 1]
    return img_array

# Predict the class of the image with a threshold
def predict_image_class(model, img_array, threshold=0.70):
    class_names = ['Cat', 'Dog', 'Panda']
    predictions = model.predict(img_array)
    predicted_probabilities = predictions[0]
    predicted_class = np.argmax(predicted_probabilities)
    confidence = predicted_probabilities[predicted_class]

    if confidence < threshold:
        return "Unknown", predicted_probabilities
    else:
        return class_names[predicted_class], predicted_probabilities

# Function to load and predict the image
def load_and_predict(img_path, threshold=0.70):
    img_array = preprocess_image(img_path)
    predicted_class, probabilities = predict_image_class(model, img_array, threshold)

    # Display the image
    img = image.load_img(img_path)
    plt.imshow(img)
    plt.axis('off')

    if predicted_class == "Unknown":
        plt.title("Predicted: Unknown")
    else:
        plt.title(f"Predicted: {predicted_class} ")

    plt.show()

    return predicted_class, probabilities

# Create a FileUpload widget
upload_widget = FileUpload()

# Display the widget
display(upload_widget)

# Handle the file upload
def on_file_upload(change):
    # Get the file contents
    uploaded_file = next(iter(change['new'].values()))
    file_contents = uploaded_file['content']

    # Save the file to disk
    with open('uploaded_image.jpg', 'wb') as f:
        f.write(file_contents)

    # Load and predict the image
    predicted_class, probabilities = load_and_predict('uploaded_image.jpg')

    # Display the prediction
    if predicted_class == "Unknown":
        print("Predicted class: Unknown")
    else:
        print(f"Predicted class: {predicted_class}")
        print(f"Probabilities: {probabilities}")

# Bind the on_file_upload function to the upload widget's 'data' event
upload_widget.observe(on_file_upload, names='value')
```
