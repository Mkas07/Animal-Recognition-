```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Define paths
base_dir = '/content/animal'
cats_dir = f'{base_dir}/cats'
dogs_dir = f'{base_dir}/dogs'
pandas_dir = f'{base_dir}/pandas'
```

```python
# Image data generator with data augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    validation_split=0.20,  # 20% of the data for validation
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Training data generator
train_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

# Validation data generator
validation_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')  # 3 classes: Cat, Dog, Panda
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=50,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)



model.save('cats_dogs_pandas_model.h5')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the trained model
model = tf.keras.models.load_model('/content/cats_dogs_pandas_modelfinal.h5')
```

```python
# Preprocess the input image
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(150, 150))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  # Rescale the image to [0, 1]
    return img_array
def predict_image_class(model, img_array, threshold=0.8):
    class_names = ['Cat', 'Dog', 'Panda']
    predictions = model.predict(img_array)
    predicted_probabilities = predictions[0]
    predicted_class = np.argmax(predicted_probabilities)
    confidence = predicted_probabilities[predicted_class]

    if confidence < threshold:
        return "Unknown", predicted_probabilities
    else:
        return class_names[predicted_class], predicted_probabilities
```

```python
import matplotlib.pyplot as plt

def load_and_predict(img_path, threshold=0.8):
    img_array = preprocess_image(img_path)
    predicted_class, probabilities = predict_image_class(model, img_array, threshold)

    # Display the image
    img = image.load_img(img_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted: {predicted_class} ({probabilities[np.argmax(probabilities)]*100:.2f}%)")
    plt.show()

    return predicted_class, probabilities

# Example usage
img_path = '/content/334289821-Baltimore_Oriole-Matthew_Plante.jpg'  # Replace with the path to your image
predicted_class, probabilities = load_and_predict(img_path)
print(f"Predicted class: {predicted_class}")
print(f"Probabilities: {probabilities}")
```
